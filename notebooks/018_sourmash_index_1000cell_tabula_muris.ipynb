```python
signature_template = '/mnt/data/sourmash_1000cell_signatures/sourmash_dna-only_trim={trim}_scaled={scale}/{cell}.signature'
```

```python
db_name_template = 'trim={trim}_scaled={scale}_dna_ksize={ksize}'

db_template = f'/mnt/data/sourmash_index_1000cell/{db_name_template}/{db_name_template}/{db_name_template}.sbt.json'
```

```python
ksize = 42
trim = 'false'
scale = 1000
cell = 'A1-D041914-3_8_M-1-1'
database = db_template.format(scale=scale, ksize=ksize, trim=trim)
query = signature_template.format(trim=trim, scale=scale, cell=cell)
! ls -lha $query $database
print()

! time sourmash search --num-results 10 --ksize $ksize --dna $query $database
```

## Now iterate over all scales, ksizes, trims, and cells

```python
scales = sourmash_compute_outputs['scaled'].unique()
scales
```

```python
ksizes = [int(x) for x in sourmash_compute_outputs['ksizes'].iloc[0].split(',')]
ksizes
```

```python
trims = 'true', 'false'
```

```python
2+2
```

```python
from itertools import product


abundance_flags = '--ignore-abundance', ''

cells = 'A1-D041914-3_8_M-1-1', 'A13-D041914-3_8_M-1-1'

csv_dir = '/mnt/data/sourmash_search_1000cell'
csv_template = '{csv_dir}/{cell}_{db_name}_{abundance}.csv'
! mkdir $csv_dir

for scale, ksize, trim, flag in product(scales, ksizes, trims, abundance_flags):
    db_name = db_name_template.format(scale=scale, ksize=ksize, trim=trim)
    
    abundance = 'ignore_abundance' if flag else "include_abundance"
    
    kwargs = dict(scale=scale, ksize=ksize, trim=trim, csv_dir=csv_dir, db_name=db_name, abundance=abundance)
    print(f'\nscale: {scale}, ksize: {ksize}, trim: {trim}, abundance: {abundance}')
    database = db_template.format(**kwargs)
    if not os.path.exists(database):
        print("\tCan't find database {os.path.basename(database)}, moving on")
        continue
    
    for cell in cells:
        kwargs['cell'] = cell
        print('\t', cell)
    
        query = signature_template.format(**kwargs)
#         print(query)
        if not os.path.exists(query):
            print("\t\tCan't find query file {os.path.basename(query)}, moving on to next one")
            continue
        csv = csv_template.format(**kwargs)
        
#         print(csv)
        if os.path.exists(csv):
            print("\t\tAlready searched! Continuing ...")
            continue
    
        ! sourmash search --output $csv --ksize $ksize $flag --dna $query $database
```
