### Human kidney transplant single cell seq

```python
human_kidney_transplant = pd.read_csv('../data/human_kidney_PRJNA431401.txt', sep='\t')
print(human_kidney_transplant.shape)
human_kidney_transplant.head()
```

### Human kidney single nucleus indrop

```python
human_kidney_nucleus = pd.read_csv('../data/human_kidney_nuclear_indrop_PRJNA464435.txt', sep='\t')
print(human_kidney_nucleus.shape)
human_kidney_nucleus.head()
```

### Zebrafish kidney and hematopoeisis

```python
zebrafish_kidney_blood = pd.read_csv('../data/zebrafish_kidney_hematopoiesis_PRJNA393429.txt', sep='\t')
print(zebrafish_kidney_blood.shape)
zebrafish_kidney_blood.head()
```

#### Add the protocol (SmartSeq2, indrop) used

```python
s = '''accession	Organism	Title
PRJNA393429	Danio rerio	Dissecting hematopoietic and renal cell heterogeneity in adult zebrafish at single cell resolution using RNA sequencing [inDrops] (Pathology, Massachusetts...)
PRJNA393430	Danio rerio	Dissecting hematopoietic and renal cell heterogeneity in adult zebrafish at single cell resolution using RNA sequencing [bulk RNA-seq] (Pathology, Massachusetts...)
PRJNA393431	Danio rerio	Dissecting hematopoietic and renal cell heterogeneity in adult zebrafish at single cell resolution using RNA sequencing [Smart-seq] (Pathology, Massachusetts...)
'''

from io import StringIO

project_to_protocol_type = pd.read_csv(StringIO(s), sep='\t', index_col=0)
print(project_to_protocol_type.shape)
project_to_protocol_type['protocol'] = project_to_protocol_type.Title.str.extract('\[([\w -]+)\]')
project_to_protocol_type['protocol'] = project_to_protocol_type['protocol'].str.rstrip('s')
project_to_protocol_type.head()
```

```python
zebrafish_kidney_blood['protocol'] = zebrafish_kidney_blood.BioProject.map(project_to_protocol_type.protocol)
zebrafish_kidney_blood['protocol'].head()
```

### Hematopoiesis continuum in humans

```python
human_continuous_hematopoiesis = pd.read_csv("../data/human_continuous_hematopoeisis_velten2017.txt", sep='\t')
human_continuous_hematopoiesis['protocol'] = 'smartseq2_quartzseq'
print(human_continuous_hematopoiesis.shape)
human_continuous_hematopoiesis.head()
```

### Mouse kidney 10x

https://www.ncbi.nlm.nih.gov/pubmed/29622724

```python
mouse_kidney_10x = pd.read_csv('../data/mouse_kidney_10x_park2018_PRJNA420723.txt', sep='\t')
mouse_kidney_10x['protocol'] = '10x_chromium'
print(mouse_kidney_10x.shape)
mouse_kidney_10x.head()
```

## Concatenate all the runs

```python
sra_data = pd.concat([human_kidney_transplant, human_kidney_nucleus, 
                      zebrafish_kidney_blood, human_continuous_hematopoiesis,
                      mouse_kidney_10x])
print(sra_data.shape)
sra_data.head()
```

```python
folder = '../data/200_download_species_data'
! mkdir -p $folder

# Add a `Run_s` column for https://github.com/pbelmann/sra-download/
sra_data['Run_s'] = sra_data['Run']
sra_data.to_csv(f'{folder}/human_zebrafish_kidney_blood.tsv', sep='\t', index=False)
```

```python
cols = ['Organism', 'protocol', 'source_name', 'tissue', 'Assay_Type', 'platform', 'BioProject']
```

```python
sra_data[cols].drop_duplicates()
```

```python
sra_data.fillna("NA").groupby(cols).size()
```

```python
sra_data['tissue'] = sra_data.apply(
    lambda x: x['tissue'] if not pd.isnull(x['tissue']) else x['source_name'], axis=1)
sra_data['protocol'] = sra_data.apply(
    lambda x: x['protocol'] if not pd.isnull(x['protocol']) else x['platform'], axis=1)
sra_data[['Organism', 'protocol', 'source_name', 'tissue', 'Assay_Type', 'platform']].drop_duplicates()

# inDrops --> indrop
sra_data['protocol'] = sra_data['protocol'].str.rstrip('s')
```

```python

s3_prefix = 's3://olgabot-maca/sra/'

sra_data['id'] = sra_data.apply(
    lambda x: '{Organism}_{protocol}_{source_name}_{Run}'.format(**x).lower().replace(' ', '_'), axis=1)
sra_data['output'] = sra_data.apply(
    lambda x: s3_prefix + '{Organism}/{protocol}/{source_name}_{BioProject}/'.format(**x).lower().replace(' ', '_'), axis=1)
sra_data[['id', 'output']].drop_duplicates()
```

## Make samples file for downloads

```python
samples = sra_data[['id', 'Run', 'output']].set_index('id')
samples = samples.rename(columns={"Run": "sra_id"})
samples.head()
```

```python

import json

folder = '../download_sra/kidney_human_zebrafish/'
! mkdir -p $folder

assert samples.index.is_unique
samples.to_csv(f'{folder}/samples.csv')

config = 	{
		"program": "../../../reflow-workflows/workflows/download_sra.rf",
		"runs_file": "samples.csv"
	}

with open(f'{folder}/config.json', 'w') as f:
    json.dump(config, f)
    
! head -n 2 $folder/samples.csv $folder/config.json
```
