```python
folder = '../data/200_download_species_data'
! mkdir -p $folder

# Add a `Run_s` column for https://github.com/pbelmann/sra-download/
sra_data['Run_s'] = sra_data['Run']
sra_data.to_csv(f'{folder}/human_zebrafish_kidney.tsv', sep='\t', index=False)
```

```python
sra_data[['Organism', 'protocol', 'source_name', 'tissue', 'Assay_Type', 'platform', 'BioProject']].drop_duplicates()
```

```python
sra_data['tissue'] = sra_data.apply(
    lambda x: x['tissue'] if not pd.isnull(x['tissue']) else x['source_name'], axis=1)
sra_data['protocol'] = sra_data.apply(
    lambda x: x['protocol'] if not pd.isnull(x['protocol']) else x['platform'], axis=1)
sra_data[['Organism', 'protocol', 'source_name', 'tissue', 'Assay_Type', 'platform']].drop_duplicates()

# inDrops --> indrop
sra_data['protocol'] = sra_data['protocol'].str.rstrip('s')
```

```python

s3_prefix = 's3://olgabot-maca/sra/'

sra_data['id'] = sra_data.apply(
    lambda x: '{Organism}_{protocol}_{source_name}_{Run}'.format(**x).lower().replace(' ', '_'), axis=1)
sra_data['output'] = sra_data.apply(
    lambda x: s3_prefix + '{Organism}/{protocol}/{source_name}_{BioProject}/'.format(**x).lower().replace(' ', '_'), axis=1)
sra_data[['id', 'output']].drop_duplicates()
```

## Make samples file for downloads

```python
samples = sra_data[['id', 'Run', 'output']].set_index('id')
samples = samples.rename(columns={"Run": "sra_id"})
samples.head()
```
