```python
def parse_dashing_dist(s):

    df = pd.read_table(StringIO(s), index_col=0)
    df = df.replace('-', np.nan)
    df = df.astype(float)

    try:
        df = df.drop(['##Names '], axis=1)
    except KeyError:
        pass

    df.columns = df.columns.str.split('.').str[0]
    df.index = df.columns

    # Use upper triangle of matrix to fill lower triangle and make a symmetric matrix
    data = np.triu(df) + np.triu(df).T

    # Make a pandas dataframe with the symmetric matrix
    df = pd.DataFrame(data, index=df.columns, 
                                                  columns=df.columns)
    df = df.fillna(1)
    return df
```

### Read the upper trianglar matrices and convert them to proper tsvs

```python
import glob
from io import StringIO
import os

import numpy as np
import pandas as pd

filenames = glob.glob(os.path.join(data_folder, "*uppertriangular"))

for filename in filenames:
    with open(filename) as f:
        s = f.read()
        df = parse_dashing_dist(s)
        tsv = filename.replace('uppertriangular', 'tsv')
        df.to_csv(tsv, sep='\t')
```
