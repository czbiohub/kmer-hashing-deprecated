```python
import pandas as pd

sourmash_compute_folder = '../sourmash/maca/facs_v8_tissue-subset_num-hashes=1000000/'
sourmash_samples = pd.read_csv(f"{sourmash_compute_folder}/samples.csv", index_col=0)
print(sourmash_samples.shape)
sourmash_samples.head()
```

```python
samples = sourmash_samples.copy()
samples = samples.drop(
    ['ksizes', 'num_hashes', 'dna', 'protein',
     'name', 'trim_low_abundance_kmers', 'output'], axis=1)
samples['name'] = samples.index

samples.head()
```

## Add dashing parameters

```python
import itertools

ksizes = 15, 21, 27, 33, 51

log2_sketch_sizes = 8, 9, 10, 11, 12, 13, 14, 15, 16

data = list(itertools.product(samples.index, ksizes, log2_sketch_sizes))

parameters = pd.DataFrame(data, columns=['id', 'ksize', 'log2_sketch_size'])
print(parameters.shape)
parameters = parameters.set_index('id')
parameters.head()
```

```python
samples_with_parameters = samples.join(parameters)

s3_prefix = 's3://olgabot-maca/dashing/bladder_kidney_spleen_subset/'
samples_with_parameters['output'] = samples_with_parameters.apply(
    lambda x: s3_prefix + 'ksize={ksize}_log2sketchsize={log2_sketch_size}/{name}'.format(**x), 
    axis=1)

print(samples_with_parameters.shape)
samples_with_parameters.head()
```
