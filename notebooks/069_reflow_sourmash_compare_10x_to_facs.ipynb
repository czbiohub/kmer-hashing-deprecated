```python
random_bladder = annotation_facs.query('tissue == "Bladder"').sample(n=150, random_state=0)
print(random_bladder.shape)
random_bladder.head()
```

```python
facs_prefix = 's3://olgabot-maca/facs/sourmash_compute_all/'
```

```python
bladder_facs = facs_prefix + random_bladder['cell_id'] + '.sig'
bladder_facs = bladder_facs.tolist()
bladder_facs[:5]
```

```python
random_lung = annotation_facs.query('tissue == "Lung"').sample(n=600, random_state=0)
print(random_lung.shape)
random_lung.head()
```

```python
lung_facs = facs_prefix + random_lung['cell_id'] + '.sig'
lung_facs = lung_facs.tolist()
lung_facs[:5]
```

```python
lung_tenx = 's3://olgabot-maca/10x/10X_P7_8/scaled=1000_num_hashes=0/possorted_genome_bam.sig'
bladder_tenx = 's3://olgabot-maca/10x/10X_P4_3/scaled=1000_num_hashes=0/possorted_genome_bam.sig'
```

```python
bladder_signatures = '|'.join(bladder_facs + [bladder_tenx])
bladder_signatures[:200]
```

```python
lung_signatures = '|'.join(lung_facs + [lung_tenx])
lung_signatures[:200]
```

```python
tenx_tissues = dict(Bladder=bladder_tenx, Lung=lung_tenx)
tenx_tissues["Bladder"]
```

```python
facs_tissues = dict(Bladder='|'.join(bladder_facs), Lung='|'.join(lung_facs))
facs_tissues['Bladder'][:500]
```

```python
import itertools

ksize = 27
molecules = 'protein', 'dna'
ignore_abundances = True, False
tissues = 'Bladder', 'Lung'

data = itertools.product(molecules, ignore_abundances, tissues, tissues)
samples = pd.DataFrame(list(data), columns=['sequence_to_compare', 'ignore_abundance', 
                                      'tenx_tissue', 'facs_tissue'])
samples['ksize'] = ksize
print(samples.shape)
samples
```

```python
def make_signatures(row):
    tenx_signatures = tenx_tissues[row.tenx_tissue]
    facs_signatures = facs_tissues[row.facs_tissue]
    return '|'.join([tenx_signatures, facs_signatures])
    
samples['signatures'] = samples.apply(make_signatures, axis=1)
samples.head()
```

```python
def make_id(row):
    name = 'tenx={tenx_tissue}_facs={facs_tissue}_molecule={sequence_to_compare}_'\
        'ksize={ksize}_ignore_abundance={ignore_abundance}'.format(**row)
    return name

samples['id'] = samples.apply(make_id, axis=1)
samples.head()
```

```python
output_prefix = 's3://olgabot-maca/10x_vs_facs/sourmash_compare/'
```

```python
samples['output'] = output_prefix + samples['id'] + '.csv'
samples.head()
```

```python
samples_cleaned = samples.set_index("id")
samples_cleaned = samples_cleaned.drop(columns=['tenx_tissue', 'facs_tissue'])
samples_cleaned.head()
```
