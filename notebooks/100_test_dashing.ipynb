```python
def tidify_simliarities(similarities):
    similarities_tidy = similarities.replace(1, np.nan)
    similarities_tidy = similarities_tidy.unstack()
    similarities_tidy = similarities_tidy.dropna()
    similarities_tidy = similarities_tidy.reset_index()
    similarities_tidy = similarities_tidy.rename(columns={
        'level_0':'cell_a', 'level_1': 'cell_b', 0: 'similarity'})
    similarities_tidy = similarities_tidy.set_index(['cell_a', 'cell_b'])
    return similarities_tidy
```

```python
dashing_similarities_tidy = tidify_simliarities(dashing_similarities)
dashing_similarities_tidy['method'] = 'dashing/HLL'
dashing_groundtruth_tidy = tidify_simliarities(dashing_groundtruth)
dashing_groundtruth_tidy['method'] = 'dashing/groundtruth'


sourmash_similarities_tidy = tidify_simliarities(sourmash_similarities)
sourmash_similarities_tidy['method'] = 'sourmash/MinHash'
sourmash_similarities_tidy.head()
```

```python
sourmash_similarities_tidy['diff'] = dashing_groundtruth_tidy.similarity - \
    sourmash_similarities_tidy['similarity']
dashing_similarities_tidy['diff'] = dashing_groundtruth_tidy.similarity - \
    dashing_similarities_tidy['similarity']


groundtruth = dashing_groundtruth_tidy['similarity']
groundtruth.name = 'ground_truth'

similarities_tidy = pd.concat([dashing_similarities_tidy, sourmash_similarities_tidy])
similarities_tidy = similarities_tidy.join(groundtruth)
print(similarities_tidy.shape)
similarities_tidy.head()
```
