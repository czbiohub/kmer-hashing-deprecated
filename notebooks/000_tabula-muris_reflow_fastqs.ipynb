```python
maca_aws = pd.read_table('aws_maca_remux_bytes_fastq.txt', sep='\s+', header=None, 
                         names=['date', 'time', 'filesize', 'suffix'])

print(maca_aws.shape)

maca_aws.head()
```

```python
maca_aws['sample_id'] = maca_aws['suffix'].str.split('/').str[-2]
maca_aws.head()
```

```python
maca_aws['read'] = maca_aws['suffix'].str.extract('_(R[12])_')
maca_aws.head()
```

```python
grouped = maca_aws.groupby(['sample_id', 'read'])
grouped
```

## Use larger of two files for each R1, R2
Use the largest file for each sample since the smaller one is just index hopping -_-

```python
# grouped.apply(lambda df: df.loc[df['filesize'] == df['filesize'].max()]).head()
```

This command takes ~5-10min

```python
maca_aws_largest_file = grouped.apply(lambda df: df.loc[df.filesize == df.filesize.max()])
maca_aws_largest_file.head()
```

```python
maca_aws_largest_file['s3_path'] = 's3://czbiohub-maca/' + maca_aws_largest_file['suffix']
maca_aws_largest_file.head()
```

```python
pd.options.display.max_colwidth = 500
```

```python
fastqs = maca_aws_largest_file.pivot_table(index='sample_id', columns='read', values='s3_path',
                              aggfunc=lambda x: '|'.join(x))
print(fastqs.shape)
fastqs.head()
```

```python
multiple_files_rows = fastqs.R1.str.contains('|', regex=False) | fastqs.R2.str.contains('|', regex=False)
print(multiple_files_rows.sum())
fastqs.loc[multiple_files_rows]
```

```python
fastqs.to_csv('fastqs.csv')
! head fastqs.csv
```

```python
sourmash_fastqs = fastqs.copy()
```

```python
sourmash_fastqs.index.name = 'id'
sourmash_fastqs = sourmash_fastqs.rename(columns={'R1':"read1", "R2":"read2"})
sourmash_fastqs['sample_id'] = sourmash_fastqs.index
sourmash_fastqs['output_signature'] = 's3://olgabot-maca/facs/sourmash/' + sourmash_fastqs['sample_id'] + '.signature'
sourmash_fastqs.head()
```
