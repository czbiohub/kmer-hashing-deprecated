```python
import pandas as pd

aws_fastqs = pd.read_table('ashley_180516_A00111_0149_AH5CM2DSXX.txt', 
                           header=None, sep='\s+', 
                           names=['date', 'time', 'bytes', 'suffix'])
aws_fastqs['read'] = aws_fastqs['suffix'].str.extract('_(R[12])_')
aws_fastqs['sample_id'] = aws_fastqs['suffix'].str.split('/').str[-2]
aws_fastqs['s3_path'] = 's3://czbiohub-seqbot/' + aws_fastqs['suffix']
aws_fastqs.head()
```

```python
aws_fastqs['sample_id'] = aws_fastqs['suffix'].str.split('/').str[-2]
aws_fastqs.head()
```

```python
fastqs = aws_fastqs.pivot_table(index='sample_id', columns='read', values='s3_path',
                              aggfunc=lambda x: '|'.join(x))
print(fastqs.shape)
fastqs.head()
```

```python
sourmash_fastqs = fastqs.copy()

sourmash_fastqs.index.name = 'id'
sourmash_fastqs = sourmash_fastqs.rename(columns={'R1':"read1", "R2":"read2"})
sourmash_fastqs['sample_id'] = sourmash_fastqs.index
sourmash_fastqs['output_signature'] = 's3://olgabot-maca/lung_cancer/sourmash/' + sourmash_fastqs['sample_id'] + '.signature'
sourmash_fastqs.head()

! mkdir -p /home/ubuntu/kmer-hashing/sourmash/lung_cancer/

sourmash_fastqs.to_csv('/home/ubuntu/kmer-hashing/sourmash/lung_cancer/samples.csv')
# sourmash_fastqs.to_csv('fastqs_sourmash.csv')
sourmash_fastqs.head()
```
