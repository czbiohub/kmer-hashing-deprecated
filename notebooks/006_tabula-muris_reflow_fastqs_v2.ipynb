```python
import numpy as np

np.arange(2000, 5001, step=500)
```

```python
parameters = {'trim_low_abundance_kmers': ('true', 'false'),
              'scaled': np.concatenate((np.arange(100, 2000, step=100), np.arange(2000, 5001, step=500)))}
parameters
```

```python
np.product([len(x) for x in parameters.values()])
```

```python
import itertools
```

### Use randomly chosen cells

```python
import numpy as np
np.random.seed(0)

rows = np.random.choice(annotations.index, size=1000, replace=False)
annotations_subset = annotations.loc[rows]
print(annotations_subset.groupby('tissue').size())

random_sample_ids = annotations_subset.index
random_sample_ids[:5]
```

```python
list(itertools.product(parameters['trim_low_abundance_kmers'], parameters['scaled']))
```

```python
fastqs.head()
```

```python
ksizes = np.arange(3, 61, step=3)
ksizes
```

```python
full_parameters = itertools.product(parameters['trim_low_abundance_kmers'], parameters['scaled'])

annotated_fastqs = fastqs.join(sample_id_with_annotation)
annotated_fastqs.index.name ='id'
annotated_fastqs = annotated_fastqs.rename(columns={'R1':"read1", "R2":"read2", 
                                                  "sample_id_with_annotation": 'name'})
null_rows = annotated_fastqs.name.isnull()
annotated_fastqs.loc[null_rows, 'name'] = annotated_fastqs.index[null_rows]
annotated_fastqs_subset = annotated_fastqs.loc[random_sample_ids]
print(annotated_fastqs_subset.shape)
annotated_fastqs_subset.index.name = 'id'
dfs = []

for trim_low_abundance_kmers, scaled in full_parameters:
    df = annotated_fastqs_subset.copy()
    suffix = f'trim={trim_low_abundance_kmers}_scaled={scaled}'
    df['output'] = f's3://olgabot-maca/facs/sourmash_dna-only_{suffix}/' + df.index + '.signature'
    df.index = df.index.map(lambda x: f'{x}_{suffix}')
    df['trim_low_abundance_kmers'] = trim_low_abundance_kmers
    df['scaled'] = scaled
    df['ksizes'] = ','.join(map(str, ksizes))
    df['protein'] = 'false'
    dfs.append(df)
    
samples = pd.concat(dfs)
print(samples.shape)
samples.head()
```
