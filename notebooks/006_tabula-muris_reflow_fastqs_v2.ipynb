```python
annotations = pd.read_csv('/mnt/data/tabula-muris/00_data_ingest/18_global_annotation_csv/annotations_facs.csv', 
                          index_col='cell')
annotations.index = annotations.index.str.replace('.', '-')
annotations['sample_id'] = annotations.index
annotations = annotations.fillna("NA")
print(annotations.shape)
annotations.head()
```

```python
maca_aws = pd.read_table('aws_maca_remux_bytes_fastq.txt', sep='\s+', header=None, 
                         names=['date', 'time', 'filesize', 'suffix'])
maca_aws['megabytes'] = maca_aws['filesize']/1e6
maca_aws['novaseq_run'] = maca_aws['suffix'].str.split('/').str[1]
maca_aws['novaseq_run_number'] = maca_aws['suffix'].str.extract('_(\d{4})_')
maca_aws['s3_path'] = 's3://czbiohub-maca/' + maca_aws['suffix']
maca_aws['sample_id'] = maca_aws['suffix'].str.split('/').str[-2]
maca_aws['sample_id_no_sample_number'] = maca_aws.sample_id.map(lambda x: '_'.join(x.split('_')[:3]))
print(maca_aws.shape)

maca_aws.head()
```

```python
bad_runs = '170907_A00111_0051_BH2HWLDMXX', '170907_A00111_0052_AH2HTCDMXX'
```

```python
sample_id_with_annotation.head()
```

```python

maca_aws = maca_aws.join(sample_id_with_annotation, on='sample_id_no_sample_number')
# maca_aws.head()
print(maca_aws.shape)
maca_aws['read'] = maca_aws['suffix'].str.extract('_(R[12])_')
maca_aws.head()
```

```python
sample_id_to_annotation = pd.Series(index=maca_aws.sample_id.values, data=maca_aws.sample_id_with_annotation.values)
print(sample_id_to_annotation.shape)
# If no annotation, keep original sample id
ind = sample_id_to_annotation.isnull()
sample_id_to_annotation[ind] = sample_id_to_annotation.index[ind].values
sample_id_to_annotation = sample_id_to_annotation.drop_duplicates()
print(sample_id_to_annotation.shape)
sample_id_to_annotation.head()
```

```python
rows = maca_aws.novaseq_run.isin(bad_runs)
maca_aws_bad_runs = maca_aws.loc[rows]
print(maca_aws_bad_runs.shape)
maca_aws_bad_runs.head()
```

```python
grouped = maca_aws_bad_runs.groupby(['sample_id', 'read'], as_index=False, group_keys=False)
grouped
```

## Use larger of two files for each R1, R2, but only for  the novaseq run: `"[AB]0005[12]"`


Use the largest file for each sample since the smaller one is just index hopping -_-

```python
%time bad_run_largest_files = grouped.apply(lambda df: df.loc[df.filesize == df.filesize.max()])
bad_run_largest_files.head()
```

Combine with good runs

```python
rows = ~maca_aws.novaseq_run.isin(bad_runs)
maca_aws_good_runs = maca_aws.loc[rows]

combined_runs = pd.concat([bad_run_largest_files, maca_aws_good_runs], ignore_index=True)
print(combined_runs.shape)
combined_runs.head()
```

```python
combined_runs.to_csv('/home/ubuntu/kmer-hashing/sourmash/maca/facs/aws_metadata.csv',index=False)
```

```python
pd.options.display.max_colwidth = 500
```

```python
fastqs = combined_runs.pivot_table(index='sample_id_no_sample_number', columns='read', values='s3_path',
                              aggfunc=lambda x: '|'.join(x))
fastqs.index.name = 'sample_id'
print(fastqs.shape)
fastqs.head()
```

```python
multiple_files_rows = fastqs.R1.str.contains('|', regex=False) | fastqs.R2.str.contains('|', regex=False)
print(multiple_files_rows.sum())
fastqs.loc[multiple_files_rows].head()
```

```python
fastqs.to_csv('fastqs.csv')
! head fastqs.csv
```

```python
sourmash_fastqs = fastqs.join(sample_id_with_annotation)

sourmash_fastqs.index.name = 'id'
sourmash_fastqs = sourmash_fastqs.rename(columns={'R1':"read1", "R2":"read2", 
                                                  "sample_id_with_annotation": 'sample_id'})
# sourmash_fastqs['sample_id'] = sourmash_fastqs.index
null_rows = sourmash_fastqs.sample_id.isnull()
sourmash_fastqs.loc[null_rows, 'sample_id'] = sourmash_fastqs.index[null_rows]
sourmash_fastqs['output_signature'] = 's3://olgabot-maca/facs/sourmash/' + sourmash_fastqs.index + '.signature'

folder = '/home/ubuntu/kmer-hashing/sourmash/maca/facs_v2/'

! mkdir -p $folder

sourmash_fastqs.to_csv(f'{folder}/samples.csv')
sourmash_fastqs.to_csv('fastqs_sourmash.csv')
sourmash_fastqs.head()
```
