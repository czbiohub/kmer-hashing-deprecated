## Iterate over multilple parameters with a smallish subset

- Add protein signatures
- ksizes = 21, 33, 51
- `sample_id` --> `name`

```python
import itertools
import pandas as pd
import numpy as np
```

```python


v5_samples = pd.read_csv('../sourmash/maca/facs_v5_1000cell_dna-only_scaled_trim_comparison/samples.csv')
v5_samples.head()
```

```python
v5_samples_scaled100 = v5_samples.query('scaled == 100')
print(v5_samples_scaled100.shape)
v5_samples_scaled100.head()
```

### Make scaled 10k samples

```python
pd.options.display.max_colwidth = 500
```

```python
v5_samples_scaled10k = v5_samples_scaled100.copy()
v5_samples_scaled10k['id'] = v5_samples_scaled10k['id'].str.replace('scaled=100', 'scaled=10000')
v5_samples_scaled10k['output'] = v5_samples_scaled10k['output'].str.replace('scaled=100', 'scaled=10000')
v5_samples_scaled10k['scaled'] = 10000
v5_samples_scaled10k.head()
```

### Make scale, trim/not trimmed parameters

```python
parameters = {'trim_low_abundance_kmers': ('true', 'false'),
              'scaled': np.concatenate((np.arange(100, 2000, step=100), np.arange(2000, 5001, step=500), 
                                        [10000]))}
parameters
```

```python
np.product([len(x) for x in parameters.values()])
```

```python
list(itertools.product(parameters['trim_low_abundance_kmers'], parameters['scaled']))
```

```python
import numpy as np

ksizes = np.arange(63, 103, step=3)
ksizes
```

### Make just fastqs to iterate over

```python
import os

fastqs = v5_samples_scaled10k[['read1', 'read2', 'name']]
fastqs['id'] = fastqs['read1'].map(lambda x: '_'.join(os.path.basename(x).split("_")[:3]))
fastqs = fastqs.set_index('id')
print(fastqs.shape)
fastqs = fastqs.drop_duplicates()
print(fastqs.shape)
fastqs.head()
```

```python
full_parameters = itertools.product(parameters['trim_low_abundance_kmers'], parameters['scaled'])

dfs = []

for trim_low_abundance_kmers, scaled in full_parameters:
    df = fastqs.copy()
    suffix = f'trim={trim_low_abundance_kmers}_scaled={scaled}'
    df['output'] = f's3://olgabot-maca/facs/sourmash_compare/ksize=63-102_{suffix}/' + df.index + '.signature'
    df.index = df.index.map(lambda x: f'{x}_{suffix}')
    df['trim_low_abundance_kmers'] = trim_low_abundance_kmers
    df['scaled'] = scaled
    df['ksizes'] = ','.join(map(str, ksizes))
    df['protein'] = 'false'
    dfs.append(df)
    
samples = pd.concat(dfs)
print(samples.shape)
samples.head()
```
