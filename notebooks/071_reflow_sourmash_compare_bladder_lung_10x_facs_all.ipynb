```python
import itertools

ksize = 27
molecules = 'protein', 'dna'
ignore_abundances = True, #False
tissues = 'Bladder', 'Lung'

data = itertools.product(molecules, ignore_abundances)
samples = pd.DataFrame(list(data), columns=['sequence_to_compare', 'ignore_abundance'])
samples['ksize'] = ksize
print(samples.shape)
samples
```

```python
samples['signatures_files'] = signatures
samples.head()
```

```python
def make_id(row):
    name = 'tenx=Bladder-Lung_facs=Bladder-Lung_molecule={sequence_to_compare}_'\
        'ksize={ksize}_ignore_abundance={ignore_abundance}'.format(**row)
    return name

samples['id'] = samples.apply(make_id, axis=1)
samples.head()
```

```python
output_prefix = 's3://olgabot-maca/10x_vs_facs/sourmash_compare/'
```

```python
samples['output'] = output_prefix + samples['id'] + '.csv'
samples.head()
```
