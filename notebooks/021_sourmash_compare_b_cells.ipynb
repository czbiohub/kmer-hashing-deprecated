```python
import pronto

cell_ontology = pronto.Ontology('https://raw.githubusercontent.com/obophenotype/cell-ontology/master/cl-basic.obo')
```

```python
b_cell = cell_ontology['CL:0000236']
b_cell
```

```python
b_cell_or_child = [x.id for x in b_cell.rchildren()]
```

```python
annotations_b_cell = annotations.query('cell_ontology_id in @b_cell_or_child')
print(annotations_b_cell.shape)
annotations_b_cell.head()
```

```python
annotations_b_cell['cell_id'] = annotations_b_cell.cell.str.replace('.', '-')
annotations_b_cell.head()
```

```python
import itertools

ksizes = 21, 27, 33, 51
sequences = 'protein', 'dna'
abundance_flags = '', '--ignore-abundance'

for ksize, seq, abundance in itertools.product(ksizes, sequences, abundance_flags):
    print(f'ksize: {ksize}, seq: {seq}, abundance: {abundance}')
    track_abundance = True if abundance else False
    csv = f'/mnt/data/tabula-muris-b-cell-compare/k{ksize}-seq={seq}-abundance={track_abundance}.csv'
    print(f'\t{csv}')
    ! time sourmash compare --quiet --csv $csv --$seq --ksize $ksize $abundance --traverse-directory /mnt/data/maca-facs-sourmash_compute_all_b_cells/
```
