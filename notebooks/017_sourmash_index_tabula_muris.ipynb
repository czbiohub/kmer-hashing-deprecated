```python
import pandas as pd

sourmash_compute_samples = pd.read_csv('../sourmash/maca/facs_compute_v7_all/samples.csv')
sourmash_compute_samples.head()
```

```python
ksizes = [int(x) for x in sourmash_compute_samples['ksizes'].unique()[0].split(',')]
ksizes
```

```python
sequence_to_compare = 'protein', 'dna'
```

```python
import itertools

data = list(itertools.product(ksizes, sequence_to_compare))
data
```

```python
samples = pd.DataFrame(data, columns=['ksize', 'sequence_to_compare'])
samples
```

```python
samples['id'] = samples.apply(lambda x: 'tabula-muris-k{ksize}-{sequence_to_compare}'.format(**x), axis=1)
samples
```

```python
pd.options.display.max_colwidth = 100
```

```python
samples['database_name'] = samples['id']
samples['signatures'] = 's3://olgabot-maca/facs/sourmash_compute_all/'
samples['output'] = 's3://olgabot-maca/facs/sourmash_index_all/' + samples['database_name']
samples
```

```python
samples_index = samples.set_index('id')
samples_index
```

```python
import json

folder = '../sourmash/maca/facs_index_v7_all'

! mkdir -p $folder
samples_index.to_csv(f'{folder}/samples.csv')

config = 	{
		"program": "../../../reflow/sourmash_index.rf",
		"runs_file": "samples.csv"
	}

with open(f'{folder}/config.json', 'w') as f:
    json.dump(config, f)
    
! head -n 3 $folder/samples.csv $folder/config.json
```
