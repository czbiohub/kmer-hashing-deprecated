````python
import pandas as pd

samples = pd.read_csv('samples_first20.csv', index_col=0)
samples
````

````python
OUTDIR = 'catted_reads'
RAWDIR = 'raw_reads'

! mkdir $OUTDIR
! mkdir $RAWDIR

for sample_id, row in samples.iterrows():
    read1s = list(row.read1.split('|'))
    read2s = list(row.read2.split('|'))
    reads = read1s + read2s
    rawdir = f'{RAWDIR}/{sample_id}'
    ! mkdir $rawdir
    for i, read in enumerate(reads):
        raw_output = f'{rawdir}/{i}.fastq.gz'
        ! aws s3 sync --dryrun $read $raw_output
        
````

Concatenate the gzipped files. [Apparently it really *is* that easy](https://stackoverflow.com/questions/8005114/fast-concatenation-of-multiple-gzip-files)

````python
OUTDIR = 'catted_reads'
RAWDIR = 'raw_reads'

for sample_id, row in samples.iterrows():
    rawdir = f'{RAWDIR}/{sample_id}'
    ! cat $rawdir/*.fastq.gz > $OUTDIR/$sample_id\.fastq.gz
````
